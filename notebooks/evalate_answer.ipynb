```python
df = df.rename(
    columns={"Питання": "question", "ВІдповідь людини": "human_answer","Не правельні відповіді": "dont_right_answer"}
)
df
```

```python
from llm import LLM
llm = LLM()
```

```python
df["llm_answer"] = df["question"].apply(
    lambda q: llm.request_to_llm(q)["answer_llm"].replace("\n", " ")
)
```

```python
from langchain_openai import OpenAIEmbeddings
import config

model_embedding = OpenAIEmbeddings(api_key=config.OPENAI_API_KEY)
```

Добавимо косинусну відстань ембедінгів у правильних та неправильних відповідей

```python
from scipy.spatial.distance import cosine

df['right_answer_cosine'] = df.apply(
    lambda row: cosine(
        model_embedding.embed_documents([row['question']])[0],
        model_embedding.embed_documents([row['llm_answer']])[0]
    ),
    axis=1
)
```

```python
df['dont_right_answer_cosine'] = df.apply(
    lambda row: cosine(
        model_embedding.embed_documents([row['question']])[0],
        model_embedding.embed_documents([row['dont_right_answer']])[0]
    ),
    axis=1
)
```

Подивимся наскільки відрізняється косинусна відстань. В косіносной відстані, чим менше показник - тим краще, тим ближче вектори один до одного

```python
df['result_cosine'] = df['dont_right_answer_cosine'] - df['right_answer_cosine']
df[['result_cosine', 'dont_right_answer_cosine', 'right_answer_cosine']]
```

```python
from langchain_community.cross_encoders import HuggingFaceCrossEncoder


model_cross_encoder = HuggingFaceCrossEncoder(model_name="intfloat/multilingual-e5-small")
```

Задача кросс енкодера - видати показник, наскільки 2 пари текстів відносяться один до одного (умовно на скільки вони відповідають парі питання-відповідь).
Чим більше показник видає ця модель тим краще.

```python
text_1 = "Скільки буде 2 + 2"
right_answer = "2 + 2 буде 4"
dont_right_answer = "в океані живуть моржі"
```

```python
right = model_cross_encoder.score([(text_1, right_answer)])
right
```

```python
dont_right = model_cross_encoder.score([(text_1, dont_right_answer)])
dont_right
```

```python
df['right_answer_cross_encoder'] = df.apply(
        lambda row: model_cross_encoder.score([(row['question'], row['llm_answer'])]),
        axis=1
)
df['dont_right_answer_cross_encoder'] = df.apply(
        lambda row: model_cross_encoder.score([(row['question'], row['dont_right_answer'])]),
        axis=1
)
```

```python
df['result_cross_encoder'] = df['right_answer_cross_encoder'] - df['dont_right_answer_cross_encoder']
df[['result_cross_encoder', 'right_answer_cross_encoder', 'dont_right_answer_cross_encoder']]
```

# Висновок, який можна зробити:

1) У нашій реалізації крос-енкодер є слабким місцем. З останньої таблиці можна побачити, що деякі відповіді мають від'ємне значення. Це означає, що модель обирала менш релевантні відповіді, які не відповідали на питання.

2) Косинусна відстань добре працює для відбору документів (від'ємних значень там немає).

3) Для подальшої реалізації рекомендується:
   - **Знайти та протестувати іншу модель крос-енкодера**.
   - **Відмовитися від використання крос-енкодера як такого**.
   - **Реалізувати власний крос-енкодер** (наприклад, через невелику LLM, яка оцінюватиме від 0 до 100, наскільки один текст відповідає іншому).
